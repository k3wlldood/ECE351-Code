1. Create the following signals with user-defined functions.  
$$h_1(t) = e^{2t}u(1 - t)$$  
$$h_2(t) = u(t - 2) - u(t - 6)$$  
$$h_3(t) = \cos{(\omega_0 t)}u(t)$$ for, $f_0 = 0.25$ Hz.  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
steps = 1e-2
t = np.arange(-10, 10 + steps, steps)
n = len(t)

def step(t):
    y = np.zeros(t.shape)
    for i in range(len(t)):
        if t[i] < 0:
            y[i] = 0
        else:
            y[i] = 1
    return y

def convolve(f1, f2):
    Nf1 = len(f1)
    Nf2 = len(f2)
    f1Extended = np.append(f1, np.zeros((1, Nf2-1)))
    f2Extended = np.append(f2, np.zeros((1, Nf1-1)))
    result = np.zeros(f1Extended.shape)
    for i in range(Nf1+Nf2-1):
        result[i] = 0
        for j in range(Nf1):
            if i-j >= 0 and i-j < Nf2:
                result[i]+=f1Extended[j]*f2Extended[i-j]
    return result
```

```python
f = 0.25
w = 2*np.pi*f

def h1(t):
    y = np.exp(2*t)*step(1-t)
    return y

def h2(t):
    y = step(t-2)-step(t-6)
    return y

def h3(t):
    y = np.cos(w*t)*step(t)
    return y
```

2. Plot the three functions in a single figure (separate subplots) from $-10 \le t \le 10$ with time steps small enough to achieve appropriate resolution. 

```python
h1 = h1(t)

plt.figure()
plt.subplot(3, 1, 1)
plt.plot(t, h1)
plt.grid()
plt.ylabel("h1(t)")
plt.xlabel("t [s]")
plt.title("Transfer Function 1")

h2 = h2(t)

plt.figure()
plt.subplot(3, 1, 2)
plt.plot(t, h2)
plt.grid()
plt.ylabel("h2(t)")
plt.xlabel("t [s]")
plt.title("Transfer Function 2")

h3 = h3(t)

plt.figure()
plt.subplot(3, 1, 3)
plt.plot(t, h3)
plt.grid()
plt.ylabel("h3(t)")
plt.xlabel("t [s]")
plt.title("Transfer Function 3")
plt.tight_layout()
plt.show()
```
